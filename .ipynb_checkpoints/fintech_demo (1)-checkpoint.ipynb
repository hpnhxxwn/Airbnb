## Prep Label

```python
df = df.query("loan_status == 'Fully Paid' or loan_status == 'Charged Off' ")
```

```python
df['loan_status'] = df.loan_status.map({"Charged Off": 1, "Fully Paid": 0})
```

```python
df.loan_status.value_counts()
```

```python
## default rate
'{:.1%}'.format(sum(df.loan_status == 1) / float(df.shape[0]))
```

## Prep out of time testing set and in-time training set

```python
df['issue_d'] = df.issue_d.apply(lambda x: x.split("-")[0])
df_test = df.query("issue_d == 'Dec' or issue_d == 'Nov' or issue_d == 'Oct'")
df_train = df.query("issue_d != 'Dec' and issue_d != 'Nov' and issue_d != 'Oct'")
df_test['train_flg'] = 0
df_train['train_flg'] = 1
df = pd.concat((df_train, df_test), axis=0, ignore_index=True)
```

## Feature Validatoin

#### Invalid feature

```python
for column_name in df.columns.values:
    
    newName = column_name.replace("_", "")
    if newName not in same:
        if newName != 'trainflg':
            df.drop([column_name], axis = 1, inplace = True )
```

```python
df.shape
```

##### NLP needed feature

```python
feature_nlp_needed = ["desc", "title"]
df.drop(feature_nlp_needed, axis = 1, inplace=True)
```

## Feature Engineering & Cleaning

#### Basic cleaning

```python
df.replace('n/a', np.nan,inplace=True)
```

```python
df.int_rate = pd.Series(df.int_rate).str.replace('%', '').astype(float)
```

```python
df.revol_util = pd.Series(df.revol_util).str.replace('%', '').astype(float)
```

```python
df.emp_length.fillna(value=0,inplace=True)
df['emp_length'].replace(to_replace='[^0-9]+', value='', inplace=True, regex=True)
df['emp_length'] = df['emp_length'].astype(int)
```

```python
from datetime import datetime
df.earliest_cr_line = pd.to_datetime(df.earliest_cr_line)
dttoday = datetime.now().strftime('%Y-%m-%d')
df.earliest_cr_line = df.earliest_cr_line.apply(lambda x: (
        np.timedelta64((x - pd.Timestamp(dttoday)),'D').astype(int))/-365)
```

#### Ordinal feature encoding 

```python
grade_dic = {"A": 1, 
            "B": 2, 
            "C": 3, 
            "D": 4, 
            "E": 5,
            "F": 6, 
            "G": 7}
```

```python
df.grade = df.grade.map(grade_dic)
```

```python
df.sub_grade = df.sub_grade.apply(lambda x: (grade_dic[x[0]] - 1) * 5 + int(x[1]))
```

#### High cardinality feature encoding

###### Zip Code - frequency encoding 

```python
df.zip_code.head()
```

```python
df.zip_code = df.zip_code.apply(lambda x: x[0:3])
```

```python
zip_code_freq = df.groupby("zip_code").size().reset_index()
```

```python
zip_code_freq.columns = ["zip_code", "zip_code_freq"]
```

```python
df = pd.merge(df, zip_code_freq, how = "left", on = "zip_code")
```

```python
df.drop("zip_code", axis = 1, inplace=True)
```

###### Addr_state - frequency encoding

```python
addr_state_freq = df.groupby("addr_state").size().reset_index()
addr_state_freq.columns = ["addr_state", "addr_state_freq"]
df = pd.merge(df, addr_state_freq, how = "left", on = "addr_state")
#df.drop("zip_code", axis = 1, inplace=True)
```

##### Employment title - frequency encoding

```python
emp_title_freq = df.groupby("emp_title").size().reset_index()
emp_title_freq.columns = ["emp_title", "emp_title_freq"]
df = pd.merge(df, emp_title_freq, how = "left", on = "emp_title")
df.drop("emp_title", axis = 1, inplace=True)
```

#### One hot encoding

```python
feat_dummy = ["term", "home_ownership", "verification_status", "purpose", "initial_list_status"]
```

```python
df_all = pd.get_dummies(df.drop(['id', 'member_id'], axis = 1))
```

```python
df_all.shape
```

##  XGBoost 

### Preliminary manually parameter tuning based on stratified train-test split

```python
from sklearn.cross_validation import StratifiedKFold
from sklearn.cross_validation import train_test_split
#Kfolds = StratifiedKFold(df_all['loan_status'], n_folds = 3, shuffle=True, random_state=2019)
```

```python
df_train = df_all.query("train_flg == 1")
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(df_train.drop("loan_status", axis =1), df_train.loan_status, 
                                                      test_size=0.3, random_state=2016, stratify = df_train.loan_status )
```

```python
dtrain = xgb.DMatrix(X_train, y_train, missing = np.NAN)
dvalid = xgb.DMatrix(X_valid, y_valid, missing = np.NAN)
```

```python
params = {"objective": "binary:logistic", 
          "booster" : "gbtree", 
          "eta": 0.05, 
          "max_depth": 6, 
          "subsample": 0.632, 
          "colsample_bytree": 0.7,
          #"colsample_bylevel": 0.6,
          "silent": 1, 
          "seed": 1441, 
          "eval_metric": "auc",
          #"gamma": 1, 
          "min_child_weight": 5} # 74453
```

```python
watchlist = [(dtrain, 'train'), (dvalid, 'eval')]
num_boost_round = 1500
gbm = xgb.train(params, dtrain, num_boost_round, evals=watchlist,\
  early_stopping_rounds= 50)
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn import linear_model, datasets
import pylab as pl
def draw_ROC(model, dtrain, dvalid, y_train, y_valid):
    probas_ = model.predict(dvalid)
    probas_1 = model.predict(dtrain)
    fpr, tpr, thresholds = roc_curve(y_valid, probas_)
    fpr_1, tpr_1, thresholds_1 = roc_curve(y_train, probas_1)
    roc_auc = auc(fpr, tpr)
    roc_auc_1 = auc(fpr_1, tpr_1)
    print "Area under the ROC curve - test: %f" % roc_auc
    print "Area under the ROC curve - train: %f" % roc_auc_1
    # Plot ROC curve
    plt.figure(figsize=(8,8))
    plt.plot(fpr, tpr, label='ROC curve - test(AUC = %0.2f)' % roc_auc, color='r')
    plt.plot(fpr_1, tpr_1, label='ROC curve - train (AUC = %0.2f)' % roc_auc_1, color='b')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC for lead score model')
    plt.legend(loc="lower right")
    plt.show()
    sns.despine()
```

### ROC curve

```python
draw_ROC(gbm, dtrain, dvalid, y_train, y_valid)
```

### Feature importance

```python
import operator
```

```python
def create_feature_map(features):
    outfile = open('xgb.fmap', 'w')
    i = 0
    for feat in features:
        outfile.write('{0}\t{1}\tq\n'.format(i, feat))
        i = i + 1
    outfile.close()
```

```python
features = list(X_train.columns)
```

```python
create_feature_map(features)
```

```python
importance = gbm.get_fscore(fmap='xgb.fmap')
importance = sorted(importance.items(), key=operator.itemgetter(1))
```

```python
df_importance = pd.DataFrame(importance, columns=['feature', 'fscore'])
df_importance['fscore'] = df_importance['fscore'] / df_importance['fscore'].sum()

plt.figure(figsize=(32, 32))
df_importance.plot()
df_importance.plot(kind='barh', x='feature', y='fscore', legend=False, figsize=(6, 10))
plt.title('XGBoost Feature Importance')
plt.xlabel('relative importance')
#plt.gcf().savefig('feature_importance_xgb.png')
```
